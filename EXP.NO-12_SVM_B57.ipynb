# MISSING VALUE TREATMENT

```python
df['glucose'].fillna(value = df['glucose'].mean(),inplace=True)
```

```python
df['education'].fillna(value = df['education'].mean(),inplace=True)
```

```python
df['heartRate'].fillna(value = df['heartRate'].mean(),inplace=True)
```

```python
df['BPMeds'].fillna(value = df['BPMeds'].mean(),inplace=True)
```

```python
df['cigsPerDay'].fillna(value = df['cigsPerDay'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
#Splitting the dependent and independent variables.
x = df.drop("TenYearCHD",axis=1)
y = df['TenYearCHD']
```

```python
x
```

# Train Test Split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

# SVM Classifier

```python
from sklearn.impute import SimpleImputer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Create an imputer object that replaces NaN values with the mean of each column
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the training data and transform both training and test data
x_train_clean = imputer.fit_transform(x_train)
x_test_clean = imputer.transform(x_test)

# Train the SVC model
svc = SVC()
svc.fit(x_train_clean, y_train)

# Evaluate the model
acc = svc.score(x_test_clean, y_test) * 100
print(f"Accuracy: {acc:.2f}%")
```
